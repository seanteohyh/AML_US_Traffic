```python
from sklearn.utils import resample,shuffle

#set the minority class to a seperate dataframe
df_1 = df_train[df_train['VEHOWN'] == 0]
#set other classes to another dataframe
other_df = df_train[df_train['VEHOWN'] == 1]  
#upsample the minority class
df_1_upsampled = resample(df_1,random_state=42,n_samples=12500,replace=True)
#concatenate the upsampled dataframe
df_upsampled = pd.concat([df_1_upsampled,other_df])
df_upsampled.head(5)
```

```python
df_upsampled.shape, df_1_upsampled.shape, df_1.shape, other_df.shape
```

```python
cols = ['HOUSEID', 'VEHOWN']
X_train = df_upsampled.drop(cols, axis=1)
y_train = df_upsampled['VEHOWN']
```

```python
#logistic regression 

estimator = linear_model.LogisticRegression()
estimator.fit(X_train, y_train)
```

```python
X_test.drop('HOUSEID', axis=1, inplace=True)
X_test.head(2)
```

```python
y_pred = estimator.predict(X_test)
```

```python
y_pred
```

```python
y_test.drop('HOUSEID', axis=1, inplace=True)
y_test.head(2)
```

```python
from sklearn import metrics

cf_matrix = metrics.confusion_matrix(y_test, y_pred)
accuracy = metrics.accuracy_score(y_test, y_pred)
precision = metrics.precision_score(1-y_test, 1-y_pred)
recall = metrics.recall_score(1-y_test, 1-y_pred)

report = f"""
Confusion Matrix:\n
{cf_matrix}\n
Accuracy:\t {accuracy}
Precision:\t {round(precision, 3)}
Recall:\t\t {round(recall, 3)} 
"""
print(report)
```

```python
import matplotlib.pyplot as plt

y_pred_prob = estimator.predict_proba(X_test)
# keep probabilities for the positive class only
y_pred_prob = y_pred_prob[:, 1]

# calculate the ROC curve
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred_prob)

# Calculate AUC
print('AUC:', metrics.roc_auc_score(y_test, y_pred_prob))

plt.plot([0, 1], [0, 1], ls = '--')
plt.plot(fpr, tpr, marker = '.')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve')
plt.show()
```

```python
clf = []
desc = []
for c in [1, 10, 100]:
    clf.append(linear_model.LogisticRegression(C = c, solver = 'liblinear'))
    desc.append('Logistic Regression with C = ' + str(c))

for d in [2, 3, 4]:
    clf.append(tree.DecisionTreeClassifier(criterion = 'entropy', max_depth = d))
    desc.append('Decision Tree with depth = ' + str(d))

for i in range(6):
    print('------------------------')
    print(desc[i])
    print('------------------------')
    for j in range(3):
        print('**using ' + (j > 0 and 'Stratified' or '') + 'KFold' + (j == 2 and ' with shuffle' or ''))
        score = model_selection.cross_val_score(clf[i], X, y, cv = kf[j])
        pred = model_selection.cross_val_predict(clf[i], X, y, cv = kf[j])
        print('        score:', score)
        print('        mean: %.3f, standard deviation: %.3f' % (np.mean(score), np.std(score)))
```
